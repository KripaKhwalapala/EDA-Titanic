## Problem 1: 
Below, calculate the mean and standard deviation for the age of the passengers on the Titanic. Once you have completed the calculations, please use a markdown box to interpret the results.


```python
Titanic=pd.read_csv("Titanic.csv")
Titanic
```

```python
m=Titanic["Age"].mean()
s=Titanic["Age"].std()

print("Mean=", m)
print("Standard deviation=", s)
```

- **Hence, mean and standard deviation for the age of the passengers on the Titanic is 29.69 and 14.52 respectively.**


The mean age of 29.69 indicates that, on average, the age of the passengers is around 29.69 years.

The standard deviation of 14.52 reflects the variability or spread of ages around the mean. A higher standard deviation suggests that the ages of the passengers are more widely dispersed from the mean age of 29.69. 



## Problem 2:
Below, calculate the median and the range for the age of the passengers on the Titanic. Once you have completed the calculations, please use a markdown box to interpret the results.


```python
med=Titanic["Age"].median()
r=Titanic["Age"].max()-Titanic["Age"].min()
print("Median=", med)
print("Range=", r)
```

- **Hence, median and range for the age of the passengers on the Titanic is 28 and 79.58 respectively.**

The median age of 28 indicates that 50% of the passengers were 28 years old or younger, and the other 50% were 28 years old or older. Unlike the mean, the median is not affected by extreme values and provides a robust measure of central tendency.


The range of 79.58 represents the difference between the maximum and minimum ages in the dataset. In this case, it suggests that the oldest passenger was 79.58 years older than the youngest passenger. The range provides a simple measure of dispersion but may be influenced by outliers.


## Problem 3:
Graph the age of the passengers using a density plot. Describe/interpret the plot in a markdown box.


```python

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.kdeplot(Titanic["Age"], fill = True)

plt.title ("Density plot of Age of passengers")
```

### Density Plot of Age

The density plot illustrates the distribution of passenger ages:

- **X-Axis (Age)**: Represents the range of ages of the passengers.
- **Y-Axis (Density)**: Represents the density of passengers within each age group, indicating how common or uncommon certain ages are among the passengers.

#### Key Observations

1. **Age Distribution**:
   The density plot shows how the ages are distributed across the passengers. The shape of the curve indicates the frequency of different age groups.


2. **Mode**:
   The peak of the density plot represents the mode, or the most common age group among the passengers. For example, if the peak occurs around 30 years, it means that there are many passengers in that age group.


3. **Spread**:
   The width of the plot indicates the spread of ages. A wider spread suggests a more diverse age range among the passengers, while a narrower spread indicates less variation in ages.


4. **Tails**:
   The tails of the density plot show the distribution of the youngest and oldest passengers. If the tails are long, it suggests that there are passengers at both very young and very old ages.
   

#### Implications

- **Demographic Insights**:
  The density plot provides valuable insights into the age demographics of the passengers. Identifying the most common age groups can help in understanding the target audience or clientele.

- **Service Planning**:
  Knowing the age distribution can assist in planning services and amenities that cater to the most common age groups. For instance, a significant number of young passengers might indicate a need for family-friendly facilities, while a larger number of older passengers might necessitate more accessibility features.

- **Marketing Strategies**:
  Marketing efforts can be tailored based on the age distribution. For example, promotional activities can be focused on the age groups that are most prevalent among the passengers.

This density plot provides a clear visualization of the age distribution among passengers, highlighting key demographic characteristics and enabling informed decision-making.

## Problem 6:
Choose measures of center and spread to analyze the Fare variable. Interpret the results in a markdown box.


```python
#measures of center 

mean_fare = Titanic['Fare'].mean()
median_fare = Titanic['Fare'].median()
print(f"Mean Fare: {mean_fare:.2f}")
print(f"Median Fare: {median_fare:.2f}")


##mode also

#measures of spread

std_dev_fare = Titanic['Fare'].std()
iqr_fare = Titanic['Fare'].quantile(0.75) - Titanic['Fare'].quantile(0.25)
range_fare = Titanic['Fare'].max() - Titanic['Fare'].min()
print(f"Standard Deviation of Fare: {std_dev_fare:.2f}")
print(f"Interquartile Range (IQR) of Fare: {iqr_fare:.2f}")
print(f"Range of Fare: {range_fare:.2f}")
```



#### Interpretation:
- The **mean fare** provides an overall average, but given the standard deviation and range, it is evident that there is significant variability in the fare amounts.
- The **median fare** being lower than the mean suggests a right-skewed distribution, where a few high fares increase the mean.
- The **standard deviation** and **IQR** both indicate considerable spread, with the IQR focusing on the middle 50% and the standard deviation giving a broader measure of dispersion.
- The **range** highlights the wide disparity between the lowest and highest fares, suggesting the presence of outliers or a diverse pricing structure.

Overall, the measures of center and spread provide a comprehensive view of the fare distribution, indicating both central tendencies and variability within the data.

## Problem 16:
Were people who paid more for a ticket more likely to survive?


```python
# Define fare groups (you can adjust the threshold as needed)

fare_threshold = Titanic['Fare'].median()  # Use median fare as the threshold
Titanic['FareGroup'] = pd.cut(Titanic['Fare'], bins=[0, fare_threshold, Titanic['Fare'].max()], labels=['Low Fare', 'High Fare'])

# Calculate survival rates for each fare group
survival_rates = Titanic.groupby('FareGroup')['Survived'].mean()

survival_rates
```

- **Hence, yes. People who paid more for a ticket were more likely to survive.**

## Problem 18: 
Create your own interesting question and find an answer.


- **How many males and females were in 1st class?**

```python
# Filter the data for passengers in 1st class
first_class_passengers = Titanic[Titanic['Pclass'] == 1]

# Count the number of males and females in 1st class
male_count = first_class_passengers[first_class_passengers['Sex'] == 'male'].shape[0]
female_count = first_class_passengers[first_class_passengers['Sex'] == 'female'].shape[0]

print("Number of males in 1st class:", male_count)
print("Number of females in 1st class:", female_count)
```

## Problem 19:
What percentage of the passengers were men? What percentage were women?


```python
# Count the number of male and female passengers
male_count = (Titanic['Sex'] == 'male').sum()
female_count = (Titanic['Sex'] == 'female').sum()

# Calculate the percentage of male and female passengers
total_passengers = len(Titanic)
male_percentage = (male_count / total_passengers) * 100
female_percentage = (female_count / total_passengers) * 100

print("Percentage of passengers who were men:", male_percentage)
print("Percentage of passengers who were women:", female_percentage)
```
